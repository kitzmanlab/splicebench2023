```python
ron_xls = pd.ExcelFile( '41467_2018_5748_MOESM6_ESM.xlsx' )
ron1 = pd.read_excel( ron_xls, 'HEK293T rep1' )
ron2 = pd.read_excel( ron_xls, 'HEK293T rep2' )
ron3 = pd.read_excel( ron_xls, 'HEK293T rep3' )
```

```python
ron1 = ron1[ [ 'mutation', 'mutation type', 'genomic position', 'AE inclusion (%)',
       'AE skipping (%)', 'full IR (%)', 'first IR (%)', 'second IR (%)',
       'other (%)' ] ]
```

```python
ron2 = ron2[ [ 'mutation', 'mutation type', 'genomic position', 'AE inclusion (%)',
       'AE skipping (%)', 'full IR (%)', 'first IR (%)', 'second IR (%)',
       'other (%)' ] ]
```

```python
ron3 = ron3[ [ 'mutation', 'mutation type', 'genomic position', 'AE inclusion (%)',
       'AE skipping (%)', 'full IR (%)', 'first IR (%)', 'second IR (%)',
       'other (%)' ] ]
```

```python
wt1 = ron1.iloc[ : 2]
```

```python
wt2 = ron2.iloc[ : 2]
```

```python
wt3 = ron3.iloc[ : 2]
```

```python
wt1
```

```python
ron1 = ron1.iloc[ 2: 1944 ]
```

```python
ron2 = ron2.iloc[ 2: 1917 ]
```

```python
ron3 = ron3.iloc[ 2: 1917 ]
```

```python
iso_cols = [ 'AE inclusion (%)', 'AE skipping (%)', 'full IR (%)', 'first IR (%)', 'second IR (%)', 'other (%)' ]
```

```python
for col in iso_cols:
    
    ron1[ 'rep1_%c ' + col ] = ron1[ col ] - wt1.iloc[ 0 ][ col ]
    ron2[ 'rep2_%c ' + col ] = ron2[ col ] - wt2.iloc[ 0 ][ col ]
    ron3[ 'rep3_%c ' + col ] = ron3[ col ] - wt3.iloc[ 0 ][ col ]
    
    ron1[ 'rep1_z' + col ] = ( ron1[ col ] - wt1.iloc[ 0 ][ col ] ) / wt1.iloc[ 1 ][ col ]
    ron2[ 'rep2_z' + col ] = ( ron2[ col ] - wt2.iloc[ 0 ][ col ] ) / wt2.iloc[ 1 ][ col ]
    ron3[ 'rep3_z' + col ] = ( ron3[ col ] - wt3.iloc[ 0 ][ col ] ) / wt3.iloc[ 1 ][ col ]
    
    ron1[ 'rep1_%fc ' + col ] = ron1[ col ] / wt1.iloc[ 0 ][ col ]
    ron2[ 'rep2_%fc ' + col ] = ron2[ col ] / wt2.iloc[ 0 ][ col ]
    ron3[ 'rep3_%fc ' + col ] = ron3[ col ] / wt3.iloc[ 0 ][ col ]
```

```python
ron1.head()
```

```python
ron_wide = pd.merge( ron1.set_index( [ 'mutation', 'mutation type', 'genomic position' ] ),
                     ron2.set_index( [ 'mutation', 'mutation type', 'genomic position' ] ),
                     how = 'inner',
                     left_index = True,
                     right_index = True,
                     suffixes = ( '_rep1', '_rep2' ) ).reset_index()
```

```python
ron_wide
```

```python
ron_wide = pd.merge( ron_wide.set_index( [ 'mutation', 'mutation type', ] ),
                     ron3.set_index( [ 'mutation', 'mutation type', ] ),
                     how = 'inner',
                     left_index = True,
                     right_index = True,
                     suffixes = ( '', '_rep3' ) ).reset_index()
```

```python
ron_wide
```

```python
ron_wide = pp.stouffers_z( ron_wide,
                               iso_cols,
                               zcol = 'z')
```

```python
zcols = [ 'zAE inclusion (%)','zAE skipping (%)','zfull IR (%)','zfirst IR (%)','zsecond IR (%)','zother (%)' ]
```

```python
import statsmodels.stats.multitest as fdr
```

```python
for col in zcols:
    
    _, adj_p, _, _ = fdr.multipletests( [ 1 - ss.norm.cdf( np.abs( z ) ) for z in ron_wide[ col ] ],
                                        method = 'fdr_bh', )
    
    ron_wide[ 'p' + col[ 1: ] ] = adj_p
```

```python
for col in iso_cols:
    
    ron_wide[ '%c ' + col ] = ron_wide[ [ c for c in ron_wide if c.endswith( '%c ' + col ) ] ].mean( axis = 1 )
```

```python
for col in iso_cols:
    
    ron_wide[ '%fc ' + col ] = ron_wide[ [ c for c in ron_wide if c.endswith( '%fc ' + col ) ] ].mean( axis = 1 )
```

```python
for col in iso_cols:
    
    ron_wide[ col + '_sdv' ] = ( ron_wide[ 'p' + col ] < .05 ) & ( ron_wide[ '%c ' + col ].abs() >= 5 )
```

```python
#if sdv in anything but exon inclusion, call SDV
ron_wide[ 'sdv' ] = ron_wide[ [ iso + '_sdv' for iso in iso_cols ] ].any( axis = 1 )
```
